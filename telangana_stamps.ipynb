How does the revenue generated from document registration vary
across districts in Telangana? List down the top 5 districts that showed the highest document registration revenue growth between FY 2019 and 2022

```python
# Filter data for FY 2019 and FY 2022
start_date_2019 = pd.to_datetime('2019-04-01')  # Assuming FY starts from April
end_date_2019 = pd.to_datetime('2020-03-31')
start_date_2022 = pd.to_datetime('2022-04-01')
end_date_2022 = pd.to_datetime('2023-03-31')

df_2019 = df[(df['month'] >= start_date_2019) & (df['month'] <= end_date_2019)]
df_2022 = df[(df['month'] >= start_date_2022) & (df['month'] <= end_date_2022)]

# Calculate total revenue for each district in FY 2019 and FY 2022
revenue_2019 = df_2019.groupby('dist_code')['documents_registered_rev'].sum()
revenue_2022 = df_2022.groupby('dist_code')['documents_registered_rev'].sum()

# Calculate revenue growth
revenue_growth = (revenue_2022 - revenue_2019) / revenue_2019 * 100
```

**Documentns Revenue District wise in fy 2019 to 2022**

```python
districts_revenue = revenue_growth
print(districts_revenue)
```

**Documentns Revenue Top 5 District wise in fy 2019 to 2022**

```python
# Get top 5 districts with highest growth
top_5_districts = revenue_growth.nlargest(5)

print("Top 5 Districts with Highest Document Registration Revenue Growth:")
print(top_5_districts)
```
